## initialize input data

- vectorize input image matrices into vectors (row-major order)

```python
# ================================================== 
# fill up the blank
#
vec_0   = im_0.ravel() #row vectorize
vec_1   = im_1.ravel()
#
# ================================================== 
```

```python
print('size of vec 0 : ', vec_0.shape)
print('size of vec 1 : ', vec_1.shape)
```

## initialize a linear layer for the neural network

- initialize the weights in a fully connected layer (zero matrix)
- dimension
  - input : length of input data
  - output : 1

```python
#linear layer, 모든 weight들이 0을 갖도록 initialize
#input : vectorize된 image의 length
size_input  = len(vec_0)
size_output = 1

# ================================================== 
# fill up the blank
# 모든 값이 0이고 1 * size_input인 matrix 만들기
weight = np.zeros((size_output, size_input))
#
# ================================================== 
```

```python
print('size of weight : ', weight.shape)
```

## define a linear layer 

- input
  - input data
  - dimension : $m \times 1$
- weight
  - weight associated with the layer of the neural network
  - dimension : $p \times m$
- return
  - output of the layer
  - dimension : $p \times 1$

```python
def layer_linear(input, weight):

# ================================================== 
# fill up the blank
#input : vector

    output = np.dot(weight, input)
#
# ================================================== 

    return output
```

## define sigmoid function 

```python
def activation_sigmoid(input):

# ================================================== 
# fill up the blank
#input : z
    output = 1/(1+np.exp(-input))
#
# ================================================== 

    return output 
```

## define forward propagation

```python
def propagation_forward(input, weight):

# ================================================== 
# fill up the blank
#input 벡터가 들어왔을 때 output prediction이 나오는 과정
    output      = layer_linear(input, weight) #linear layer output
    prediction  = activation_sigmoid(output)
#
# ================================================== 

    return prediction 
```

## define the loss function

- cross entropy between the ground truth and the prediction
- cross entropy : $-y \log(h) - (1-y) \log(1-h)$
  - $y$ : grount truth
  - $h$ : prediction
- output dimension : $1 \times 1$

```python
def compute_loss(prediction, label):

# ================================================== 
# fill up the blank
# y = label, h= prediction
# -y log(h) = -label*np.log(prediction)
# -(1-y)log(l-h) = -(1-label)*np.log(1-prediction)
    loss = -label*np.log(prediction) -(1-label)*np.log(1-prediction)
#
# ================================================== 

    return loss
```

## compute the gradient of the loss with respect to the model parameter

```python
def compute_gradient(input, prediction, label):

# ================================================== 
# fill up the blank
#
    residual    = prediction - label #(fw​(xi​)−yi​) = (prediction - label)
    gradient    = residual * input#(fw​(xi​)−yi​)xi​​
#
# ================================================== 

    return gradient
```

## gradient descent algorithm

- hyper-parameters

```python
number_iteration    = 500
learning_rate       = 0.01
```

```python
loss_iter   = np.zeros(number_iteration)
loss_iter_0 = np.zeros(number_iteration)
loss_iter_1 = np.zeros(number_iteration)

pred_iter_0 = np.zeros(number_iteration)
pred_iter_1 = np.zeros(number_iteration)
```

## run the gradient descent algorithm

```python
for i in range(number_iteration):

# ================================================== 
# fill up the blank
#
    pred_0  = propagation_forward(vec_0, weight) #def propagation_forward(input, weight):
    loss_0  = compute_loss(pred_0, label_0) #def compute_loss(prediction, label):
    grad_0  = compute_gradient(vec_0, pred_0, label_0) #def compute_gradient(input, prediction, label):

    pred_1  = propagation_forward(vec_1, weight)
    loss_1  = compute_loss(pred_1, label_1)
    grad_1  = compute_gradient(vec_1, pred_1, label_1)

    loss    = (loss_0 + loss_1)/2 #average
    grad    = (grad_0 + grad_1)/2
    weight  = weight - learning_rate * grad
#
# ================================================== 

    loss_iter_0[i]  = loss_0 
    loss_iter_1[i]  = loss_1
    loss_iter[i]    = loss

    pred_iter_0[i]  = pred_0
    pred_iter_1[i]  = pred_1
```
